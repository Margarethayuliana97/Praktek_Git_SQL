## **SQL statement untuk menemukan jumlah buyer per country**

Menemukan jumlah buyer per country menggunakan SQL command COUNT dan GROUP BY

```python
sql4 ="""SELECT Country, count(buyerid) as total_buyer
FROM public.user_tab
Group by country;"""
buyer_per_country = pd. read_sql_query(sql4, conn)
buyer_per_country
```

## **SQL statement untuk menemukan jumlah total item_views untuk setiap shop**

Menemukan jumlah total item_views untuk setiap shop menggunakan JOIN, COUNT dan GROUP BY

```python
sql5 ="""SELECT t1.shopid, count(t2.item_views) as total_views
FROM public.user_tab t1 JOIN public.performance_tab t2
ON t1.shopid = t2.shopid GROUP BY t1.shopid;"""
item_views_shop = pd. read_sql_query(sql5, conn)
item_views_shop
```
